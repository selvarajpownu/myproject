```python
import nltk
from nltk.corpus import words, stopwords
from nltk.metrics import edit_distance
import string
```

```python
nltk.download('words')
nltk.download('stopwords')
```

```python
# Get a list of valid words and normalize casing
valid_words = words.words()
first_20_words = valid_words[:20]
normalized_words = [word.lower() for word in valid_words]
unique_normalized_words = list(set(normalized_words))
```

```python
# Create a list of stop words
stop_words = set(stopwords.words('english') + list(string.punctuation))
```

```python
# Define a function to get the correct term
def get_correct_term(term):
    valid_words_subset = valid_words[:20000]
    distances = {word: edit_distance(term, word) for word in valid_words_subset}
    sorted_distances = sorted(distances.items(), key=lambda x: x[1])
    return sorted_distances[0][0]
```

```python
# Create a set from the list of valid words
valid_words_set = set(valid_words)
```

```python
# Define a function for spelling correction
def correct_spelling_in_sentence(sentence):
    tokenized_sentence = nltk.word_tokenize(sentence.lower())
    corrected_terms = [term if term in valid_words_set else get_correct_term(term) for term in tokenized_sentence]
    corrected_sentence = ' '.join(corrected_terms)
    return corrected_sentence
```

```python
# 1 Test the function
input_sentence = "The new abacos is great"
output_sentence = correct_spelling_in_sentence(input_sentence)
print("Input Sentence:", input_sentence)
print("Corrected Sentence:", output_sentence)
```
