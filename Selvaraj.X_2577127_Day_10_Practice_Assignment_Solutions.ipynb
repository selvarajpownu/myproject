```python
# Define a year as a "Superman year" whose films feature more Superman characters than Batman. 
#How many years in film history have been Superman years?

superman_batman_films = cast[(cast['character'] == 'Superman') | (cast['character'] == 'Batman')]

character_counts = superman_batman_films.groupby(['year', 'character'])['title'].count().unstack(fill_value=0)

superman_years = character_counts[character_counts['Superman'] > character_counts['Batman']]

num_superman_years = len(superman_years)

print(f"The number of years in film history that have been Superman years: {num_superman_years}")
```

```python
# How many years have been "Batman years", with more Batman characters than Superman characters?

superman_batman_films = cast[(cast['character'] == 'Superman') | (cast['character'] == 'Batman')]

# Group by 'year' and 'character' to count the number of Superman and Batman characters in each year
character_counts = superman_batman_films.groupby(['year', 'character'])['title'].count().unstack(fill_value=0)

batman_years = character_counts[character_counts['Batman'] > character_counts['Superman']]

# Count the number of "Batman years"
num_batman_years = len(batman_years)

print(f"The number of years in film history that have been Batman years: {num_batman_years}")
```

```python
# Plot the number of actor roles each year and the number of actress roles each year over the history of film.

roles_per_year = cast.groupby(['year', 'type'])['title'].count().unstack(fill_value=0)

# Plotting
plt.figure(figsize=(10, 6))
roles_per_year.plot(kind='line', xlabel='Year', ylabel='Number of Roles', title='Number of Actor and Actress Roles Each Year')
plt.legend(['Actor', 'Actress'])
plt.show()
```

```python
# Plot the number of actor roles each year and the number of actress roles each year, but this time as a kind='area' plot.

plt.figure(figsize=(10, 6))
roles_per_year.plot(kind='area', xlabel='Year', ylabel='Number of Roles', title='Number of Actor and Actress Roles Each Year')
plt.legend(['Actor', 'Actress'])
plt.show()
```

```python
# Plot the difference between the number of actor roles each year and the number of actress roles each year over the history of film.

role_difference = roles_per_year['actor'] - roles_per_year['actress']

plt.figure(figsize=(10, 6))
role_difference.plot(kind='line', xlabel='Year', ylabel='Difference (Actor - Actress)', title='Difference Between Actor and Actress Roles Each Year')
plt.show()
```

```python
# Plot the fraction of roles that have been 'actor' roles each year in the history of film.



# Calculate the total number of roles each year
total_roles_per_year = roles_per_year['actor'] + roles_per_year['actress']

# Calculate the fraction of roles that are 'actor' roles each year
fraction_actor_roles = roles_per_year['actor'] / total_roles_per_year

# Plotting the fraction
plt.figure(figsize=(10, 6))
fraction_actor_roles.plot(kind='line', xlabel='Year', ylabel='Fraction of Actor Roles', title='Fraction of Roles That Are "Actor" Roles Each Year')
plt.show()
```

```python
# Plot the fraction of supporting (n=2) roles that have been 'actor' roles each year in the history of film.

supporting_roles = cast[cast['n'] == 2]

# Group by 'year' and 'type' to count the number of supporting actor and actress roles in each year
supporting_roles_per_year = supporting_roles.groupby(['year', 'type'])['title'].count().unstack(fill_value=0)

# Calculate the fraction of supporting roles that are 'actor' roles each year
fraction_supporting_actor_roles = supporting_roles_per_year['actor'] / (supporting_roles_per_year['actor'] + supporting_roles_per_year['actress'])

plt.figure(figsize=(10, 6))
fraction_supporting_actor_roles.plot(kind='line', xlabel='Year', ylabel='Fraction of Actor Roles (n=2)', title='Fraction of Supporting (n=2) Roles That Are "Actor" Roles Each Year')
plt.show()
```

```python
# Build a plot with a line for each rank n=1 through n=3, 
# where the line shows what fraction of that rank's roles were 'actor' roles for each year in the history of film.


# Group by 'year', 'n', and 'type' to count the number of actor and actress roles for each rank each year
roles_by_rank_and_year = cast.groupby(['year', 'n', 'type'])['title'].count().unstack(fill_value=0)

# Calculate the fraction of roles that are 'actor' roles for each rank each year
fraction_actor_roles_by_rank = roles_by_rank_and_year['actor'] / (roles_by_rank_and_year['actor'] + roles_by_rank_and_year['actress'])

# Plotting
plt.figure(figsize=(10, 6))

# Plot a line for each rank (n=1, n=2, n=3)
for n in [1, 2, 3]:
    # Filter data for the specific rank (n) and create a DataFrame
    data_for_rank = fraction_actor_roles_by_rank[roles_by_rank_and_year.index.get_level_values('n') == n]
    
    # Plot the line
    plt.plot(data_for_rank.index.get_level_values('year'), data_for_rank.values, label=f'n={n}')

plt.xlabel('Year')
plt.ylabel('Fraction of Actor Roles')
plt.title('Fraction of Roles That Are "Actor" Roles by Rank Each Year')
plt.legend(title='Rank')
plt.show()
```
